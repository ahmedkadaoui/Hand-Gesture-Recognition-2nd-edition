# Kaggle challenge : Hand Gesture Recognition 2nd edition 

# IMPORTS

```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt

import csv

from keras.layers import Dense,Conv2D,Flatten,MaxPool2D,Dropout
from keras.models import Sequential

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

np.random.seed(1)
```

# DATA PREPROCESSING 

```python
#Loading the data 

#train data
train_csv = pd.read_csv(r'C:\Users\ahmed\Desktop\labs\train.csv')
```

```python
#test data
test_csv = pd.read_csv(r'C:\Users\ahmed\Desktop\labs\test.csv')
```

```python
#Getting file names (same as ids' in our case) , and labels 

filenames = [fname for fname in map(str , train_csv['id'].tolist())]
labels = train_csv['label'].tolist()
```

```python
print('file names: ', filenames ,'\n \n', '\nlables :', labels) 
```

```python
train_csv.head()
```

```python
 train_csv.info()
```

```python
#modeling our data 
#train data

train_images=[]
train_labels=[]

for index, row in train_csv.iterrows():    
        pixels = row[2:] 
        pixels = np.array(pixels, dtype='float').reshape((211,271))
        label = row[1]
        train_images.append(pixels)
        train_labels.append(label)
        
       
```

```python
train_labels = pd.get_dummies(train_labels).values
train_images = np.array(train_images)
```

```python
train_images
```

```python
#test data 

test_images=[]

for index, row in test_csv.iterrows():    
        pixels = row[1:] 
        pixels = np.array(pixels, dtype='float').reshape((211,271))
        test_images.append(pixels)
```

```python
test_images=np.array(test_images)
```

```python
test_images
```

```python
train_images=np.reshape(train_images,(-1,211, 271,1))
```

```python
print(train_images.shape)
```

```python
x_train,x_val,y_train,y_val = train_test_split(train_images,train_labels,random_state=1)
```

```python
# a glance at our train data 

plt.imshow(train_images[0])
```

# Building our model

```python
model= Sequential()
model.add(Conv2D(kernel_size=(3,3), filters=32, activation='tanh',input_shape=x_train.shape[1:]))
model.add(MaxPool2D(2,2))
model.add(Conv2D(filters=30,kernel_size = (3,3),activation='tanh'))
model.add(MaxPool2D(2,2))


model.add(Flatten())

model.add(Dense(32,activation='relu'))
model.add(Dropout(0.4))

model.add(Dense(16,activation = 'softmax'))
    
model.compile(
              loss='categorical_crossentropy', 
              metrics=['acc'],
              optimizer='adam'
             )
```

```python
x_train.shape[1:]
```

```python
model.summary()
```

# Training the model

```python
epochS= 15
batch_Size=32

history = model.fit(x_train,y_train,epochs=epochS,batch_size=batch_Size,validation_data=(x_val,y_val))
```

we reached 90% accuracy , we'll take it for now :)

```python
#Visualizing Training vs Validation loss


loss_train = history.history['loss']
loss_val = history.history['val_loss']
epochs = range(1,16)
plt.plot(epochs, loss_train, 'g', label='Training loss')
plt.plot(epochs, loss_val, 'b', label='validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

The loss is decreasing in both training and validation , thats means the model we built is actually learning !

```python
# #Visualizing Training vs Validation accuracy

loss_train = history.history['acc']
loss_val = history.history['val_acc']
epochs = range(1,16)
plt.plot(epochs, loss_train, 'g', label='Training accuracy')
plt.plot(epochs, loss_val, 'b', label='validation accuracy')
plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

The accuracy is increasing meaning our model is getting better at predicting ! 

# Evaluating the model

```python
predictions=model.predict(x_val)
```

```python
#confusion matrix 

y_pred=model.predict(x_val) 
y_pred=np.argmax(y_pred, axis=1)
y_val=np.argmax(y_val, axis=1)
cm = confusion_matrix(y_val, y_val)
print(cm)
```

A diagonal confusion matrix means all predictions made by our model were correct
